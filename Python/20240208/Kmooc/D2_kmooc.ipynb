### csv 파일 불러오기 실습

```python
import csv
file_name = '한국장학재단_대학별 장학금 수혜 현황.csv'

fd = open(file_name, 'r', encoding= 'cp949')

data = []
rd = csv.reader(fd)
for line in rd :
    data.append(line)
data[:5]
```

```python
col = data[0]
cnt = 0

search = input('검색할 학교명 :')
for line in data[1:] :
    if search in line[col.index('학교명')] :
        cnt += 1
        print(str(cnt) + '.')
        for i,name in enumerate(col) :
            print(f'{name} : {line[i]}')
        print('=' * 40)

print(f"총 {cnt}건 검색되었습니다.")
```
