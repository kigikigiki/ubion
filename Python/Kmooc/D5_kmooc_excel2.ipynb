```python
import openpyxl
workbook = openpyxl.load_workbook("서울특별시 가로쓰레기통 현황_202106.xlsx", data_only=True)
wb1 = workbook['Sheet1']
```

```python
for row in wb1.rows :
    for cell in row:
        print(cell.value, end = " ")
    print()
```

```python
location = wb1['D']
sel_data = []
for data in location :
    if data.value == '삼청로' :
        sel_row = data.coordinate[1:]
        pick = 'E' + sel_row
        sel_data.append(wb1[pick].value)
for each in sel_data :
    print (each)
```
