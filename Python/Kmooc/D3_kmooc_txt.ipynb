```python
file = open('TheLittlePrince.txt', 'r', encoding='cp949')
text = file.read()
print(text)
```

```python
word_list = text.split()

count_list = []
for word in word_list :
    count = word_list.count(word)
    count_list.append([word, count])

print(count_list)
```

```python
sorted_list = sorted(count_list, key = lambda x : x[1], reverse=True)
print(sorted_list)
```

```python
word_dict = {}
for word, frequency in sorted_list :
    if word not in word_dict :
        word_dict[word] = frequency
        print(word,frequency)
```

```python
# 텍스트 파일 불러오기
file_name = '먼 후일.txt'
fp = open(file_name, 'r', encoding= 'utf-8')

# 파일 읽기
data = fp.readlines()
fp.close()

print(data)
```

```python
for line in data :
    print(line, end='')
```

```python
fp2 = open(file_name.split('.')[0] + "_작업본.txt", "w")

cnt = 0
where = []

for i,line in enumerate(data) :
    if "잊었노라" in line :
        cnt += 1
        where.append(i)
        fp2.write(str(i)+ ' : ' + line)

fp2.close()

print(f"'잊었노라'는 {where}줄에 총 {cnt}번 존재합니다.")
```
