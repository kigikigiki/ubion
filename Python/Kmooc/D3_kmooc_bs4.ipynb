```python
import requests

# 네이버 뉴스 페이지 url
url = 'https://news.naver.com/'

#네이버 뉴스 페이지 요청, 응답 저장
response = requests.get(url)
print(response.text)
```

```python
from bs4 import BeautifulSoup

html = response.text

soup = BeautifulSoup(html, 'html.parser')

hdline_titles = soup.find_all('div', {'class' :'cjs_t'})

print(hdline_titles)
#for title in hdline_titles :
#    print(title.get_text(strip = True))
```

```python
# 크롤링할 웹사이트 설정
url = 'https://eiec.kdi.re.kr/policy/materialList.do'

# resp = requests.get(url) <- SSL 인증서 관련 오류
resp = requests.get(url, verify=False)

soup1 = BeautifulSoup(resp.text)

top_news = []

print(soup1)
```

```python
# 크롤링할 태그 조건 설정
strong = soup1.select('.list_txt > p')
for n in strong :
    top_news.append(n.get_text().strip())

# 크롤링 결과 확인
print(top_news)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = top_news)
```

```python
df['0'] = '제목'
```

```python
df.drop(
    ['0'],
    axis = 1,
    inplace=True
)
```
