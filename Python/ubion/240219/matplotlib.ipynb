```python
## 우리나라의 총 술소비량
## 그래프 표시
## 총술소비량이 가장 많은 국가는 어디인가?

drink['총술소비량'] = drink['맥주소비량'] + drink['위스키소비량'] + drink['와인소비량']
```

```python
drink.sort_values('총술소비량', ascending=False, inplace = True)
```

```python
drink.reset_index(drop = True, inplace=True)
```

```python
drink['순위'] = drink.index+1
flag = drink['국가'] == 'South Korea'
```

```python
drink[flag]
```

```python
# 총술소비량이 상위 10위까지 데이터를 시각화
x = drink['국가'].head(10)
y = drink['총술소비량'].head(10)

plt.bar(
    x, y
)

plt.show()
```

```python
## 총술소비량 막대 그래프 표시
## 우리나라의 막대의 색상을 변경하고 주석 달기

x = drink.index
y = drink['총술소비량']

# South Korea의 인덱스 값 추출
korea_index = drink.loc[drink['국가'] == 'South Korea',].index[0]
# South Korea의 총술소비량
korea_total = drink.loc[drink['국가'] == 'South Korea', '총술소비량']

# 그래프 사이즈
plt.figure(figsize = (14, 8))

bar_list = plt.bar(x,y)

# 한국의 막대를 다른 색으로 변경
bar_list[korea_index].set_color('red')

## 그래프에 주석 설정
plt.annotate(
    f'South Korea : {korea_index + 1}',
    xy = (korea_index, korea_total),
    xytext= (korea_index + 10, korea_total + 100),
    arrowprops = dict(
        facecolor = 'r',   # 화살표 내부의 색을 지정
        edgecolor = 'b',    # 화살표 외각의 색을 지정
        headwidth = 10,     # 화살표 머리의 너비
        headlength = 10,    # 화살표 머리의 길이
        width = 5,         # 화살표 꼬리의 너비
        shrink = 0.05       # 화살표와 주석사이의 거리
    )
)

plt.show()
```

```python
## 총술소비량 막대 그래프 표시
## 우리나라의 막대의 색상을 변경하고 주석 달기


x = drink.index
y = drink['도수']

# South Korea의 인덱스 값 추출
korea_index = drink.loc[drink['국가'] == 'South Korea',].index[0]
# South Korea의 총술소비량
korea_total = drink.loc[drink['국가'] == 'South Korea', '도수']

# 그래프 사이즈
plt.figure(figsize = (14, 8))

bar_list = plt.bar(x,y)

# 한국의 막대를 다른 색으로 변경
bar_list[korea_index].set_color('red')

## 그래프에 주석 설정
plt.annotate(
    f'South Korea : {korea_index + 1}',
    xy = (korea_index, korea_total),
    xytext= (korea_index + 10, korea_total + 1),
    arrowprops = dict(
        facecolor = 'r',   # 화살표 내부의 색을 지정
        edgecolor = 'b',    # 화살표 외각의 색을 지정
        headwidth = 10,     # 화살표 머리의 너비
        headlength = 10,    # 화살표 머리의 길이
        width = 5,         # 화살표 꼬리의 너비
        shrink = 0.05       # 화살표와 주석사이의 거리
    )
)

plt.show()
```
