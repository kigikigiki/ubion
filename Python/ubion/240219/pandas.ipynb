### pandas와 DataFrame 의 차이

- pandas -> Lib
    - 내장되어 있는 data가 없음
    - pandas 내부에 있는 함수를 사용할 때, data를 입력해줘야함

- DataFrame -> Class
    - read_csv
    - data가 존재함
    - DataFrame의 함수를 사용할 때, 인자값으로 data를 입력하지 않음


```python
from datetime import datetime

time_data = '2024-02-19'

# 문자열을 시계열 데이터로 변경
time = datetime.strptime(time_data, '%Y-%m-%d')

print(type(time_data))
print(type(time))
```

```python
## 시계열에서 특정 부분을 추출하여 문자열로 변환
time.strftime('%A')
```

### datetime에서 사용되는 format

- %Y : 년도를 전체 표시 (2024)
- %y : 년도를 축약해서 표시 (24)
- %b : 축약된 월의 이름('Mar')
- %B : 월의 이름('March')
- %m : 월을 숫자로 표시('2')
- %d : 일을 숫자로 표시 (19)
- %H : 24시를 기준으로 시를 표시
- %I : 12시를 기준으로 시를 표시
- %p : AM/PM을 표시
- %M : 분을 표시
- %S : 초를 표시
- %a : 요일의 이름을 축약해서 표시 (Mon)
- %A : 요일의 이름을 표시 (Monday)
- %w : 요일을 숫자로 표시 (0 - 6 : 일요일 - 토요일)

### 연습문제
1. csv 폴더에 있는 Sale Record.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Order Priority' 컬럼을 제거
3. 컬럼의 이름을 변경 -> ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 기존의 인덱스는 제거하고 인덱스를 재지정
6. 파생변수 생성
    - 총단가
    - 총원가
    - 총이윤

```python
sale_record = pd.read_csv('../../../../csv/Sales Records.csv')
```

```python
sale_record
```

```python
sale_record.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Order Priority'], axis=1 ,inplace=True)
```

```python
sale_record
```

```python
sale_record.columns = ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
sale_record
```

```python
sale_record.sort_values(['대륙', '국가'], ascending=True, inplace=True)
```

```python
sale_record
```

```python
sale_record.reset_index(drop = True, inplace = True)
```

```python
sale_record
```

```python
sale_record['총단가'] = sale_record['판매개수'] * sale_record['단가']
```

```python
sale_record['총원가'] = sale_record['판매개수'] * sale_record['원가']
sale_record['총이윤'] = sale_record['총단가'] - sale_record['총원가']
```

```python
sale_record
```

```python
sale_record
```

```python
sale_record_1 = sale_record.copy()
sale_record_1
```

```python
sale_record_1.groupby('대륙').sum()
```

```python
sale_record_1 = sale_record_1.loc[:,['대륙', '총이윤']]
sale_record[['대륙', '총이윤']].groupby("대륙").sum().sort_values('총이윤')
```

```python
sale_record_1.groupby('대륙').sum().sort_values('총이윤')
```

```python
sale1 = sale_record[['국가', '총이윤']].groupby("국가").sum().sort_values('총이윤' , ascending=False)
```

```python
#case1 파생변수
sale1 = sale1.reset_index()
sale1['순위'] = sale1.index + 1

flag = sale1['국가'] == 'South Korea'
sale1[flag]
```

```python
#case2 반복문
korea_rank = 1

for i in sale1.index :
    if i == 'South Korea' :
        break
    korea_rank += 1

print(korea_rank)
```

```python
#case3 파생변수
sale1['순위'] = range(1, len(sale1) + 1)

sale1.loc['South Korea', '순위']
```

```python
#case4 list 함수
list(sale1.index).index('South Korea') + 1
```

```python
## 대륙, 국가로 그룹화
sale = sale_record[['대륙', '국가', '총이윤']].groupby(['대륙', '국가']).sum()
```

```python
sale.stack()
```

```python
sale4 = pd.pivot_table(
    sale_record,
    index = ['대륙','국가'],
    columns='상품종류',
    values = '총이윤',
    aggfunc = ['sum','mean'],
    margins = True
)
```

```python
sale4['sum', 'Baby Food']
```

```python
sale4.loc['Asia', ('sum', 'Baby Food')]
```

```python
## groupby 에서 연산식이 2개 이상인 경우
sale5 = sale_record[['대륙', '국가', '상품종류','총이윤']].groupby(['대륙','국가','상품종류']).agg(['sum','mean'])
```

```python
sale5.unstack()
```
