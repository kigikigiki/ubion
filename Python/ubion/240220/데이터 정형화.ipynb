```python
def custom_replace (data, origin, new):
    # 첫번째 매개변수는 변경시킬 문자열
    # 두번째 매개변수는 기존 문자열
    # 세번째 매개변수는 기존의 문자열을 변경할 문자열
    result = ''
    for i in data :
        if i == origin:
            result += new
        else :
            result += i
    return result
```

```python
custom_replace ('abcde', 'a', 'A')
```

```python
def custom_replace2 (data, origin, new) :
    result = []
    for i in data:
        if i == origin :
            result.append(new)
        else :
            result.append(i)
    return result
```

```python
data = ['a', 'b', 'a', 'd']
custom_replace2(data,'a','A')
```

### 결측치를 다른 값들로 대체
1. 상품명에 따라 가격이 고정값을 이룬다.
2. 결측치를 상품명에 맞춰서 해당 상품의 결측치가 아닌 상품 가격들의 평균값을 대체
    - ex) 상품A인 데이터 중에 item_price가 결측치가 아닌 데이터들의 평균 값을 구하고 해당 값을 결측치에 대입

```python
def fill_price(_df, _name):
    ## _df를 copy()
    result = _df.copy()
    flag = result['item_name'] == _name
    flag_null = result['item_price'].isna()
    
    ## flag와 flag_null의 부정이 모두 만족하는 item_price의 평균
    mean_data = result.loc[flag & ~flag_null, 'item_price'].mean()
    ## 위에서 구해진 평균 데이터를 flag와 flag_null을 모두 만족하는 item_price에 대입
    result.loc[flag % flag_null, 'item_price'] = mean_data
    
    print(f"{_name} 작업완료")
    return result
```

```python
name_list = set(df['item_name'])
for name in name_list :
    fill_price(df, name)
```

```python
# item_name 중복값을 제거한 _list1
_list = df['item_name'].to_list()
_list1 = []
for i in _list :
    if i not in _list1 :
        _list1.append(i)
        
for j in _list1:
    flag = df['item_name'] == j
    flag_null = df['item_price'].isna()
    data = df.loc[flag & ~flag_null, 'item_price'].mean()
    df2.loc[flag & flag_null, 'item_price'] = data
df2
```
